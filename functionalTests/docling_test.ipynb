```python
import requests
import json
```

```python
url = "http://localhost:8000/debug/parse-pdf"
files = {'file': open('../data/testPDFs/MacLeod et al. - 2022 - A self-driving laboratory advances the Pareto front for material properties.pdf', 'rb')}

response = requests.post(url, files=files)
data = response.json()

print("Title detected:", data['metadata_extracted']['title'])
print("Sections found:", list(data['sections'].keys()))
```

```python
with open('data.json', 'w') as f:
    json.dump(data, f)
```
